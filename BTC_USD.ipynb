```python
from sklearn.metrics import accuracy_score, f1_score
import pandas as pd
import numpy as np
```

```python
price_test = pd.read_csv('./Bitcoin_price_test.csv', index_col=0)
conditions = [
    ((price_test['Close'] - price_test['Open']) < -100),
    ((price_test['Close'] - price_test['Open']) > -100) & ((price_test['Close'] - price_test['Open']) < 100),
    ((price_test['Close'] - price_test['Open']) > 100)
    ]
values = [-1, 0, 1]
price_test['label'] = np.select(conditions, values)
```

```python
price_test
```

```python
def always_the_same(price_test):
    pred_n = np.array([-1]*len(price_test))
    pred_s = np.array([0]*len(price_test))
    pred_p = np.array([1]*len(price_test))
    label = price_test['label']
    df = pd.DataFrame(
        {'accuracy': [accuracy_score(label, pred_n), accuracy_score(label, pred_s), accuracy_score(label, pred_p)],
         'macro f1': [f1_score(label, pred_n, average='macro'), f1_score(label, pred_s, average='macro'),
                     f1_score(label, pred_p, average='macro')],
         'micro f1': [f1_score(label, pred_n, average='micro'), f1_score(label, pred_s, average='micro'),
                     f1_score(label, pred_p, average='micro')]},
         index=['negative', 'stable', 'positive'])
    print(df)
    
always_the_same(price_test)
```

```python
def by_prop(price_train, price_test):
    label = np.array(price_test['label'])
    train_labels = np.array(price_train['label'])
    p = [np.count_nonzero(train_labels == -1) / len(train_labels), 
         np.count_nonzero(train_labels == 0) / len(train_labels), 
         np.count_nonzero(train_labels == 1) / len(train_labels)]
    pred = np.random.choice([-1, 0, 1], len(label), p=p)
    return accuracy_score(label, pred), f1_score(label, pred, average='macro'), f1_score(label, pred, average='micro')
    
accs = []
f11 = []
f12 = []
for i in range(1000):
    acc, f1a, f1i = by_prop(price_train, price_test)
    accs.append(acc)
    f11.append(f1a)
    f12.append(f1i)
accs = np.array(accs)
f11 = np.array(f11)
f12 = np.array(f12)
```

```python
pd.DataFrame([[accs.mean(), f11.mean(), f12.mean()]], index=['By prob'], 
             columns=['accuracy', 'macro f1', 'micro f1'])
```
